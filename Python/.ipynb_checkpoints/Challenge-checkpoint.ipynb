```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the csv data which is the list of cities with the highest number of waiting childrens in Japan
df = pd.read_csv('./data/waiting-child.csv')
```

```python
df.head()
```

```python
#Show the scatter plot
plt.style.use('default')
fig,ax = plt.subplots()
scatter_plot = ax.scatter(
    df['%eldery'],
    df['%waiting-workingF'],
    s=df["#female working fulltime(20-45)"]/100,
    c=df["#female working fulltime(20-45)"],
    cmap=plt.cm.Pastel1,
    edgecolor="#ECEFF4",
    linewidth=0.5
)
cbar = fig.colorbar(scatter_plot)
cbar.set_label("population in the city (unit:1000)")
ax.set_xlabel("aging rate in the city")
ax.set_ylabel("waiting children rate in the city")
ax.set_title("Scatter plot of aging ratio & waiting child ratio, of the cities with the highest waiting children")
plt.show()
```

```python
#limit the data to the cities which have more than 10000 female full time workers
df10000 = df.loc[df['#female working fulltime(20-45)'] >= 10000]
df10000.head()
```

```python
#show the scatter plot again
plt.style.use('default')
fig,ax = plt.subplots()
scatter_plot = ax.scatter(
    df10000['%eldery'],
    df10000['%waiting-workingF'],
    s=df10000["#female working fulltime(20-45)"]/100,
    c=df10000["#female working fulltime(20-45)"],
    cmap=plt.cm.Pastel1,
    edgecolor="#ECEFF4",
    linewidth=0.5
)
cbar = fig.colorbar(scatter_plot)
cbar.set_label("# of full time working female in the city")
ax.set_xlabel("aging rate in the city")
ax.set_ylabel("waiting children rate over working female in the city")
ax.set_title("Scatter plot of aging ratio & waiting child ratio, of the cities with the highest waiting children")
for _, row in df10000[['city-E','%eldery','%waiting-workingF']].iterrows():
    xy = row[['%eldery','%waiting-workingF']]
    xytext = xy + (0.01,0)
    ax.annotate(row['city-E'],xy,xytext)
plt.show()
```

```python
#limit the data to Tokyo only 
tokyodf = df10000.loc[df10000['state'] == 'Tokyo']
```
