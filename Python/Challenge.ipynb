```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/taiki_jido.csv')
```

```python
df.head()
```

```python
plt.style.use('default')
fig,ax = plt.subplots()
scatter_plot = ax.scatter(
    df['%eldery'],
    df['%waiting'],
    s=df["population(thousand)"],
    c=df["population(thousand)"],
    cmap=plt.cm.Pastel1,
    edgecolor="#ECEFF4",
    linewidth=0.5
)
cbar = fig.colorbar(scatter_plot)
cbar.set_label("population in the city (unit:1000)")
ax.set_xlabel("aging rate in the city")
ax.set_ylabel("waiting children rate in the city")
ax.set_title("Scatter plot of aging ratio & waiting child ratio, of the cities with the highest waiting children")
# for _, row in df[['city-E','%eldery','%waiting']].iterrows():
#     xy = row[['%eldery','%waiting']]
#     xytext = xy
#     ax.annotate(row['city-E'],xy,xytext)
plt.show()
```

```python
#limiting to Tokyo only
tokyodf = df.loc[df['state'] == 'Tokyo']
tokyodf.head()
```

```python
plt.style.use('default')
fig,ax = plt.subplots()
scatter_plot = ax.scatter(
    tokyodf['%eldery'],
    tokyodf['%waiting'],
    s=tokyodf["population(thousand)"],
    c=tokyodf["population(thousand)"],
    cmap=plt.cm.Pastel1,
    edgecolor="#ECEFF4",
    linewidth=0.5
)
cbar = fig.colorbar(scatter_plot)
cbar.set_label("population in the city (unit:1000)")
ax.set_xlabel("aging rate in the city")
ax.set_ylabel("waiting children rate in the city")
ax.set_title("Scatter plot of aging ratio & waiting child ratio, of the cities with the highest waiting children")
for _, row in tokyodf[['city-E','%eldery','%waiting']].iterrows():
    xy = row[['%eldery','%waiting']]
    xytext = xy + (0.01,0)
    ax.annotate(row['city-E'],xy,xytext)
plt.show()
```
